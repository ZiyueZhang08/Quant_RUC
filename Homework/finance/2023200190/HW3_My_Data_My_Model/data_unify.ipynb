```python
import pandas as pd
```

下面对其他三位同学的二手房数据进行格式统一

```python
df_esf_gumei_origin = pd.read_excel('data_esf_gumei.xlsx') # 这里调用组员的数据
```

```python
df_esf_gumei_origin_storage1 = df_esf_gumei_origin[['价格','每平方米价格']]
df_esf_gumei_origin_storage2 = df_esf_gumei_origin[['地铁站']]
df_esf_gumei_origin['拼接'] = df_esf_gumei_origin['小区'] +" " +df_esf_gumei_origin['小区地址']
df_esf_gumei_fragment = df_esf_gumei_origin.drop(columns=['价格','每平方米价格','小区','小区地址','地铁站'])
df_esf_gumei = pd.concat([df_esf_gumei_fragment,df_esf_gumei_origin_storage2,df_esf_gumei_origin_storage1],axis=1)
df_esf_gumei['num'] = 0
df_esf_gumei['require'] = None
df_esf_gumei
```

```python
import re

information_split = []
for s in df_esf_gumei['户型']:
    try:
        room_match1 = re.match(r"(\d+)室(\d+)厅", s) 
        bed_room_number = int(room_match1.group(1))
        living_room_number = int(room_match1.group(2))
        information_split.append([bed_room_number,living_room_number])
    except:
        information_split.append([None,None])
df_information = pd.DataFrame(information_split)

area_split = []
for s in df_esf_gumei['面积']:
    try:
        area_match = re.search(r"(\d+\.\d+|\d+)", s)
        room_match2 = float(area_match.group(1))
        area_split.append([room_match2])
    except:
        area_split.append([None])
df_area = pd.DataFrame(area_split)

layer_split = []
for s in df_esf_gumei['朝向']:
    try:
        layer_match = re.search(r"(.+?)\（共", s) 
        all_layer_match = re.search(r"共(\d+)层", s)
        layer = layer_match.group(1).strip()
        all_layer = int(all_layer_match.group(1))
        layer_split.append([layer,all_layer])
    except:
        layer_split.append([None,None])
df_layer = pd.DataFrame(layer_split)
        
location_split = []
for s in df_esf_gumei['拼接']:
    try:
        split_list = re.split(r'[\n ]', s,maxsplit=2)
        while len(split_list) < 3:
            split_list.append(None)
        location_split.append(split_list)
    except:
        location_split.append([None,None,None])
df_location = pd.DataFrame(location_split)
        

time_split = []
for s in df_esf_gumei['建造年代']:
    try:
        time_match = re.search(r'\d+', s)
        time_match = int(time_match.group())
        time_split.append([time_match])
    except:
        time_split.append([None])
df_time = pd.DataFrame(time_split)
        

distance_to_subway = []
for s in df_esf_gumei['地铁站']:
    try:
        distance = int(re.search(r'约(\d+)', s).group(1))
        distance_to_subway.append([distance])
    except:
        distance_to_subway.append([None]) 
df_distance = pd.DataFrame(distance_to_subway)

price_split = []
for s in df_esf_gumei['价格']:
    try:
        price_total = int(re.search(r'\d+',s).group())*10000
        price_split.append([price_total])
    except:
        price_split.append([None]) 
df_price = pd.DataFrame(price_split)

price_per = []
for s in df_esf_gumei['每平方米价格']:
    try:
        price_square = int(re.search(r'\d+',s).group())
        price_per.append([price_square])
    except:
        price_per.append([None])
df_price_per = pd.DataFrame(price_per)


df_esf_gumei_standard = pd.concat([df_esf_gumei['名称'],df_information,df_area,df_layer,df_esf_gumei['楼层'],df_time,df_location,df_esf_gumei['require'],df_distance,df_price,df_price_per],axis=1)
df_esf_gumei_standard
```

```python
df_esf_xujiahui_origin = pd.read_excel('data_esf_xujiahui.xlsx') # 这里调用组员的数据
df_esf_xujiahui_origin = df_esf_xujiahui_origin.drop(columns=['Unnamed: 0'])
df_esf_xujiahui_origin
```

```python
information_split = []
for s in df_esf_xujiahui_origin['Basic_Info_Raw']:
    # 初始化所有字段为None（默认不符合格式）
    bed_room_number = None
    living_room_number = None
    room_match2 = None
    layer = None
    all_layer = None
    direction_match = None
    time_match = None
    name = None

    try:
        split_list = s.split('|')
        
        room_hall = split_list[0].strip() 
        room_match1 = re.match(r"(\d+)室(\d+)厅", room_hall) 
        bed_room_number = int(room_match1.group(1))
        living_room_number = int(room_match1.group(2))
        
        room_area = split_list[1].strip()
        area_match = re.search(r"(\d+\.\d+|\d+)", room_area)
        room_match2 = float(area_match.group(1))
        
        floor_str = split_list[2].strip()
        layer_match = re.search(r"(.+?)\（共", floor_str) 
        all_layer_match = re.search(r"共(\d+)层", floor_str)
        layer = layer_match.group(1).strip()
        all_layer = int(all_layer_match.group(1))
        
        direction = split_list[3].strip()
        direction_match = direction 
        
        time = split_list[4].strip()
        time_match = re.search(r'\d+', time)
        time_match = int(time_match.group())
        
        name = split_list[5].strip()
    
    except Exception as e:
        pass

    info_group = [bed_room_number,living_room_number,room_match2,layer,all_layer,direction_match,time_match,name]
    information_split.append(info_group)
df_information_split = pd.DataFrame(information_split)

location_split = []
for s in df_esf_xujiahui_origin['Detail_Address']:
    try:
        split_list = re.split(r'[\n ]', s,maxsplit=2)
        while len(split_list) < 2:
            split_list.append(None)
        location_split.append(split_list)
    except:
        location_split.append([None,None,None])
df_location_split = pd.DataFrame(location_split)

characteristic_split = []
for s in df_esf_xujiahui_origin['Metro_Distance']:
    try:
        distance = int(re.search(r'约(\d+)', s).group(1))
        characteristic_split.append([distance])
    except:
        characteristic_split.append([None])
df_characteristic_split = pd.DataFrame(characteristic_split)

price_split = []
for s in df_esf_xujiahui_origin['Total_Price']:
    try:
        price_total = int(re.search(r'\d+',s).group())*10000
        price_split.append([price_total])
    except:
        price_split.append([None]) 
df_price = pd.DataFrame(price_split)

price_per = []
for s in df_esf_xujiahui_origin['Unit_Price']:
    try:
        price_square = int(re.search(r'\d+',s).group())
        price_per.append([price_square])
    except:
        price_per.append([None])
df_price_per = pd.DataFrame(price_per)

df_esf_xujiahui_standard = pd.concat([df_esf_xujiahui_origin['Title'],df_information_split,df_location_split,df_esf_xujiahui_origin['House_Labels'],df_characteristic_split,df_price,df_price_per],axis=1)
df_esf_xujiahui_standard

```

```python
df_esf_zhangjiang_origin = pd.read_excel("data_esf_zhangjiang.xlsx") # 这里调用组员的数据
df_esf_zhangjiang_origin
```

```python
information_split = []
for s in df_esf_zhangjiang_origin['户型']:
    try:
        room_match1 = re.match(r"(\d+)室(\d+)厅", s) 
        bed_room_number = int(room_match1.group(1))
        living_room_number = int(room_match1.group(2))
        information_split.append([bed_room_number,living_room_number])
    except:
        information_split.append([None,None])
df_information = pd.DataFrame(information_split)

layer_split = []
for s in df_esf_zhangjiang_origin['楼层']:
    try:
        layer_match = re.search(r"(.+?)\（共", s) 
        all_layer_match = re.search(r"共(\d+)层", s)
        layer = layer_match.group(1).strip()
        all_layer = int(all_layer_match.group(1))
        layer_split.append([layer,all_layer])
    except:
        layer_split.append([None,None])
df_layer = pd.DataFrame(layer_split)

distance_to_subway = []
for s in df_esf_zhangjiang_origin['地铁']:
    try:
        distance = int(re.search(r'约(\d+)', s).group(1))
        distance_to_subway.append([distance])
    except:
        distance_to_subway.append([None]) 
df_distance = pd.DataFrame(distance_to_subway)

location_split = []
for s in df_esf_zhangjiang_origin['位置']:
    try:
        split_list = re.split(r'[\n ]', s,maxsplit=2)
        while len(split_list) < 2:
            split_list.append(None)
        location_split.append(split_list)
    except:
        location_split.append([None,None,None])
df_location = pd.DataFrame(location_split)

df_esf_zhangjiang_origin['总价(万元)'] = df_esf_zhangjiang_origin['总价(万元)']*10000

df_esf_zhangjiang_standard = pd.concat([df_esf_zhangjiang_origin['标题'],df_information,df_esf_zhangjiang_origin['面积(㎡)'],df_layer,df_esf_zhangjiang_origin['朝向'],df_esf_gumei_origin['建造年代'],df_esf_zhangjiang_origin['小区'],df_location,df_esf_zhangjiang_origin['标签'],df_distance,df_esf_zhangjiang_origin['总价(万元)'],df_esf_zhangjiang_origin['单价(元/㎡)']],axis=1)
df_esf_zhangjiang_standard

column = ['描述','室数','厅数','面积','层级','总层数','朝向','始建年份','小区名称','路段','位置','房产证登记年份','地铁站','总价格','每平方米价格']
df_esf_zhangjiang_standard.columns = column
df_esf_zhangjiang_standard.loc[df_esf_zhangjiang_standard['路段'] != '张江', '路段'] = '张江'
        
```

```python
df_esf_shibo = pd.read_excel('data_esf_shibo.xlsx') # 这里调用组员的数据
df_esf_shibo_standard = df_esf_shibo.drop(columns=['出售人','特征1','特征2','特征3','特征4'])
df_esf_shibo_standard

column = ['描述','室数','厅数','面积','层级','总层数','朝向','始建年份','小区名称','路段','位置','房产证登记年份','地铁站','总价格','每平方米价格']
df_esf_shibo_standard.columns = column
df_esf_gumei_standard.columns = column
df_esf_zhangjiang_standard.columns = column
df_esf_xujiahui_standard.columns = column

df_esf = pd.concat([df_esf_shibo_standard,df_esf_gumei_standard,df_esf_zhangjiang_standard,df_esf_xujiahui_standard],axis=0)
df_esf.to_excel('data_esf.xlsx',index=False)
```

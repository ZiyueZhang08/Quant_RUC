```python
import pandas as pd
import re
```

下面对其他三位同学的租房数据进行格式统一

```python
df_zu_gumei = pd.read_excel('data_zu_gumei.xlsx') # 这里调用组员的数据
df_storage = df_zu_gumei[['价格']]
df_zu_gumei = df_zu_gumei.drop(columns=['价格'])
df_zu_gumei = pd.concat([df_zu_gumei,df_storage],axis=1)

distance_subway = []
for s in df_zu_gumei['地铁站']:
    try:
        distance = int(re.search(r'约(\d+)', s).group(1))
        distance_subway.append(distance)
    except:
        distance_subway.append(None)
df_distance_subway_split = pd.DataFrame(distance_subway)

location_split = []
for s in df_zu_gumei['小区']:
    try:
        split = s.split('-')
        location_split.append(split)
    except:
        location_split.append([None,None,None])
df_location_split = pd.DataFrame(location_split)

information_split = []
for s in df_zu_gumei['户型']:
    try:
        room_match = re.match(r"(\d+)室(\d+)厅",s)
        bed_room_number = int(room_match.group(1))
        living_room_number = int(room_match.group(2))
        information_split.append([bed_room_number,living_room_number])
    except:
        information_split.append([s,s])
df_information_split = pd.DataFrame(information_split)

area_split = []
for s in df_zu_gumei['面积']:
    try:
        room_area_match = re.search(r'(\d+\.\d+|\d+)',s)
        room_area_number = float(room_area_match.group())
        area_split.append([room_area_number])
    except:
        area_split.append([None])
df_area_split = pd.DataFrame(area_split)

price_split = []
for s in df_zu_gumei['价格']:
    try:
        price_match = re.search(r'\d+', s)
        price_number = int(price_match.group())
        price_split.append([price_number])
    except:
        price_split.append([None])
df_price_split = pd.DataFrame(price_split)

style_split = []
for s in df_zu_gumei['出租方式']:
    try:
        s1 = s.replace(':', '')
        s2 = s1.strip()
        style_split.append([s2])
    except:
        style_split.append([None])
df_style_split = pd.DataFrame(style_split)

df_zu_gumei = pd.concat([df_zu_gumei['名称'],df_style_split,df_information_split,df_area_split,df_zu_gumei['朝向'],df_location_split,df_distance_subway_split,df_price_split],axis=1)
df_zu_gumei
```

```python
df_zu_xujiahui = pd.read_excel('data_zu_xujiahui.xlsx') # 这里调用组员的数据
df_zu_xujiahui = df_zu_xujiahui.drop(columns=['Labels','Price_Unit','Unnamed: 0'])
df_zu_xujiahui['二级地址']= '徐家汇'

information_split = []
for s in df_zu_xujiahui['Basic_Info_Raw']:
    try:
        split = s.split('|')
        
        room_hall = split[1]
        room_match = re.match(r"(\d+)室(\d+)厅",room_hall)
        bed_room_number = int(room_match.group(1))
        living_room_number = int(room_match.group(2))

        room_area = split[2]
        room_area_match = re.search(r'(\d+\.\d+|\d+)',room_area)
        room_area_number = float(room_area_match.group())
        
        info_group = [split[0],bed_room_number,living_room_number,room_area_number,split[3]]
        information_split.append(info_group)
    except:
        information_split.append([None,None,None,None,None])
df_information_split = pd.DataFrame(information_split)

distance_subway = []
for s in df_zu_xujiahui['Metro_Distance']:
    try:
        distance = int(re.search(r'约(\d+)', s).group(1))
        distance_subway.append(distance)
    except:
        distance_subway.append(None)
df_distance_subway_split = pd.DataFrame(distance_subway)


df_zu_xujiahui = pd.concat([df_zu_xujiahui['Title'],df_information_split,df_zu_xujiahui['Community_Name'],df_zu_xujiahui['二级地址'],df_zu_xujiahui['Detail_Address'],df_distance_subway_split,df_zu_xujiahui['Price']],axis=1)
df_zu_xujiahui


```

```python
df_zu_zhangjiang = pd.read_excel('data_zu_zhangjiang.xlsx') # 这里调用组员的数据
information_split = []
for s in df_zu_zhangjiang['户型']:
    try:
        room_match = re.match(r"(\d+)室(\d+)厅",s)
        bed_room_number = int(room_match.group(1))
        living_room_number = int(room_match.group(2))
        information_split.append([bed_room_number,living_room_number])
    except:
        information_split.append([s,s])
df_information_split = pd.DataFrame(information_split)

location_split = []
for s in df_zu_zhangjiang['区域']:
    try:
        split = s.split('-')
        location_split.append(split)
    except:
        location_split.append([None,None,None])
df_location_split = pd.DataFrame(location_split)

distance_subway = []
for s in df_zu_zhangjiang['地铁']:
    try:
        distance = int(re.search(r'约(\d+)', s).group(1))
        distance_subway.append(distance)
    except:
        distance_subway.append(None)
df_distance_subway_split = pd.DataFrame(distance_subway)

df_zu_zhangjiang = pd.concat([df_zu_zhangjiang['标题'],df_zu_zhangjiang['租赁方式'],df_information_split,df_zu_zhangjiang['面积(㎡)'],df_zu_zhangjiang['朝向'],df_location_split,df_distance_subway_split,df_zu_zhangjiang['价格(元/月)']],axis=1)
df_zu_zhangjiang


```

```python
df_zu_shibo = pd.read_excel('data_zu_shibo.xlsx')
df_zu_shibo = df_zu_shibo.drop(columns=['特征1','特征2','特征3'])
df_zu_shibo
```

```python
column = ['描述','租赁方式','室数','厅数','面积','朝向','一级地址','二级地址','三级地址','地铁站','价格']
df_zu_shibo.columns = column
df_zu_gumei.columns = column
df_zu_zhangjiang.columns = column
df_zu_xujiahui.columns = column

data_zu = pd.concat([df_zu_shibo,df_zu_gumei,df_zu_zhangjiang,df_zu_xujiahui],axis=0)
data_zu.to_excel('data_zu.xlsx',index = False)

df_zu_shibo.to_excel('df_zu_shibo.xlsx',index = False)
df_zu_gumei.to_excel('df_zu_gumei.xlsx',index = False)
df_zu_xujiahui.to_excel('df_zu_xujiahui.xlsx',index = False)
df_zu_zhangjiang.to_excel('df_zu_zhangjiang.xlsx',index = False) # 保存各个同学的标准数据
```

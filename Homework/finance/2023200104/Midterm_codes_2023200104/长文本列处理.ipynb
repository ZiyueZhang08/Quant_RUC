```python
# 定义创建特征的函数
def create_keyword_features(df, column_name, keywords, prefix):
    """
    为 DataFrame 中的指定列和关键词列表创建二元特征。
    """
    if column_name not in df.columns:
        print(f"警告：列 '{column_name}' 不在 DataFrame 中，跳过。")
        return df
        
    df_copy = df.copy()
    
    # 填充 NaN 以便 .str 操作
    s = df_copy[column_name].fillna('').astype(str)
    
    print(f"  正在为 '{column_name}' 创建特征...")
    new_cols_count = 0
    for keyword in keywords:
        new_col_name = f"{prefix}_{keyword}"
        # 使用 regex=False 进行精确匹配，可能稍微快一点
        df_copy[new_col_name] = s.str.contains(keyword, regex=False).astype(int) 
        new_cols_count += 1
    print(f"    - 新增 {new_cols_count} 列。")
    
    return df_copy

print("函数 'create_keyword_features' 已定义。") 

# 步骤 1: 对 '周边配套' 和 '交通出行' 进行分词与词频统计
import pandas as pd
import jieba # 用于中文分词
from collections import Counter
import re

print("开始进行分词与词频统计...")

# 定义一个函数来获取词频
def get_word_frequency(series_list, top_n=100, custom_stopwords=None):
    """
    对合并后的 Pandas Series 列表进行分词，统计词频，并返回 top_n 词语。
    
    参数:
    - series_list: 包含 Pandas Series 的列表 (例如 [train_df['col'], test_df['col']])
    - top_n: 返回最高频的 N 个词
    - custom_stopwords: 一个包含要忽略的词语的 set (可选)
    
    返回:
    - 一个包含 (词语, 频率) 元组的列表
    """
    if custom_stopwords is None:
        custom_stopwords = set()

    # 合并所有文本，并处理 NaN
    full_text = pd.concat(series_list).fillna('').astype(str)
    # 使用 jieba 分词
    words = []
    for text in full_text:
        # 移除非中文字符和数字，保留基础标点可能有助于断句，但这里简化处理
        text = re.sub(r'[^\u4e00-\u9fa5]+', ' ', text) 
        seg_list = jieba.cut(text)
        # 过滤掉停用词、单字和纯空格
        words.extend([word for word in seg_list if word.strip() and len(word) > 1 and word not in custom_stopwords])
    # 统计词频
    word_counts = Counter(words)

    # 返回最高频的 N 个词
    return word_counts.most_common(top_n)

# 定义一个更全面的自定义停用词集合
stopwords = {
    # ---------------- Baisc Stopwords ----------------
    # --- Punctuation (mostly handled by regex, but good to have) ---
    '，', '。', '、', '；', '：', '？', '！', '“', '”', '‘', '’', '（', '）', '【', '】', '《', '》', '——', '……',
    # --- Pronouns ---
    '我', '你', '他', '她', '它', '们', '自己', '我们', '你们', '他们', '她们', '它们', '彼此', '大家', '个人',
    # --- Determiners / Quantifiers ---
    '这', '那', '该', '此', '这里', '那里', '这些', '那些', '什么', '多少', '几', '个', '本', '某', '每', '各', '全', '整',
    # --- Conjunctions ---
    '和', '或', '与', '且', '而', '但', '则', '就', '才', '也', '还', '并', '并且', '或者', '但是', '然而', '因为', '所以', '由于', '因此', '如果', '那么', '即使', '虽然',
    # --- Prepositions ---
    '在', '从', '向', '往', '于', '比', '被', '把', '对', '给', '为', '为了', '关于', '对于', '根据', '除了', '沿着', '通过',
    # --- Adverbs ---
    '很', '非常', '最', '都', '挺', '极其', '十分', '有点', '稍微', '更加', '越', '只', '不', '没', '没有', '未', '否', '曾', '已经', '正在', '将', '即将', '常常', '经常', '有时', '偶尔', '再次', '一再', '依然', '仍然', '必须', '应当', '应该', '可能', '大概', '也许',
    # --- Particles / Auxiliaries ---
    '的', '地', '得', '了', '着', '过', '吧', '吗', '呢', '啊', '哦', '嗯', '之', '乎', '者', '也', '矣', '哉',
    # --- Common Verbs (often auxiliary) ---
    '是', '有', '为', '做', '来', '去', '说', '看', '走', '到', '使', '让', '要', '想', '觉得', '认为', '表示', '进行', '成为',
    # --- Numbers (Words, often less useful) ---
    '一', '二', '三', '四', '五', '六', '七', '八', '九', '十', '百', '千', '万', '亿',

    # ---------------- Domain/Context Specific Stopwords ----------------
    # --- User's original list ---
    '介绍', '小区', '房子', '距离', '项目', '位于', '周边', '配套', '交通', '出行',
    '设施', '附近', '以及', '生活', '可以', '满足', '提供', '方便', '等',
    '公里', '米', '左右', '就是',

    # --- Added domain/context words ---
    '步行', '大约', '需要', '分钟', '小时', '即可', '业主', '房东', '租客', '租户',
    '出租', '出售', '房屋', '房源', '户型', '楼盘', '地址',
    '情况', '信息', '详情', '欢迎', '咨询', '联系',
    '各种', '多个', '一些', '一条', '一家', '一个', '几条', '几家',
    '包括', '日常', '需求', '齐全', '便利', '发达', '完善',
    '门口', '出口', '入口', '路口', '站', '线路', '号线', # '号线' 作为一个整体移除可能更好
    '之类', '的话', '来说',
    '即可到达', '举步可达', '直线距离', # Common phrases
    '等处', '等等', # Variations of '等'
    '本房', '此房', # Referencing the property itself
    '业主诚心', '诚心出售', '诚心出租', # Common phrases, might not add value
    '近', '出', '通往', # Verbs related to location/transportation that are too general
    '购物', # Often too general, specific names like '超市', '商场' are better
    '餐饮', # Similar to '购物'
    '娱乐', # Similar to '购物'
}

# print(f"停用词总数: {len(stopwords)}")
# print(list(stopwords)[:20]) # 打印前 20 个看看

try:
    # 准备 '周边配套' 的 Series 列表
    series_list_peitao = []
    if 'train_price_df' in locals() and train_price_df is not None: series_list_peitao.append(train_price_df['周边配套'])
    if 'test_price_df' in locals() and test_price_df is not None: series_list_peitao.append(test_price_df['周边配套'])
    # 准备 '交通出行' 的 Series 列表
    series_list_jiaotong = []
    if 'train_price_df' in locals() and train_price_df is not None: series_list_jiaotong.append(train_price_df['交通出行'])
    if 'test_price_df' in locals() and test_price_df is not None: series_list_jiaotong.append(test_price_df['交通出行'])
    if series_list_peitao:
        print("\n--- '周边配套' 高频词 (Top 50) ---")
        freq_peitao = get_word_frequency(series_list_peitao, top_n=50, custom_stopwords=stopwords)
        print(freq_peitao)
    else:
        print("\n无法获取 '周边配套' 词频，相关 DataFrame 未定义。")

    if series_list_jiaotong:
        print("\n--- '交通出行' 高频词 (Top 50) ---")
        freq_jiaotong = get_word_frequency(series_list_jiaotong, top_n=50, custom_stopwords=stopwords)
        print(freq_jiaotong)
    else:
        print("\n无法获取 '交通出行' 词频，相关 DataFrame 未定义。")

except NameError as e:
    print(f"错误：{e}。DataFrame 未定义。请确保已成功加载数据。")
except KeyError as e:
    print(f"错误：找不到列 {e}。")

print("\n词频统计完成。请根据以上词频选择关键词。")
```

```python

# 步骤 2: 基于关键词创建二元特征 (仅 Price 数据集)
print("\n开始基于关键词创建二元特征 (仅 Price)...")
# --- 定义关键词 (保持不变，或根据需要调整) ---
keywords_peitao = [
    '医院', '公园', '银行', '商场', '市场', '幼儿园', '学校', '超市', 
    '购物广场', '社区', '休闲', '医疗', '百货', '菜场', '三甲', '体育', 
    '酒店', '步行街', '便利店', '森林公园' 
]
keywords_jiaotong = [
    '地铁', '公交', '公交站', '高速', '机场', '自驾', '大道', '轻轨', 
    '隧道', '快速路', '高架', '高铁', '枢纽站'
]
try:
    # ***** 修改这里：只包含 Price 的 DataFrame *****
    dataframes_to_process = {
        'train_price_df': train_price_df,
        'test_price_df': test_price_df
    }
    for df_name, df_obj in dataframes_to_process.items():
        if df_obj is None:
            print(f"\n跳过 {df_name}，因为 DataFrame 未加载。")
            continue
        print(f"\n--- 处理 {df_name} ---")
        # 处理 '周边配套'
        processed_df = create_keyword_features(df_obj, '周边配套', keywords_peitao, '配套')
        # 处理 '交通出行'
        processed_df = create_keyword_features(processed_df, '交通出行', keywords_jiaotong, '交通')
        # 将处理后的 DataFrame 重新赋值给原变量名
        globals()[df_name] = processed_df 
        print(f"  - {df_name} 新形状: {processed_df.shape}")
    print("\n关键词特征创建完成 (仅 Price 数据集)。")
except NameError as e:
    print(f"错误：{e}。DataFrame 或 'create_keyword_features' 函数未定义。")
    print("请确保已成功加载数据并定义了函数。")
except Exception as e:
    print(f"处理过程中发生错误: {e}")

if 'train_price_df' in locals() and train_price_df is not None:
    print("\n--- 验证 train_price_df 中的新特征 (部分示例) ---")
    
    example_row = train_price_df[
        train_price_df['周边配套'].astype(str).str.contains('医院', na=False) & 
        train_price_df['交通出行'].astype(str).str.contains('地铁', na=False)
    ]
    if not example_row.empty:
        cols_to_show_verify = ['周边配套', '配套_医院', '交通出行', '交通_地铁']
        cols_to_show_verify = [col for col in cols_to_show_verify if col in example_row.columns]
        if cols_to_show_verify:
             print(example_row[cols_to_show_verify].head())
        else:
             print("要验证的列 ('配套_医院', '交通_地铁') 不存在。")
    else:
        print("未找到包含 '医院' 和 '地铁' 的样本进行验证。")
```

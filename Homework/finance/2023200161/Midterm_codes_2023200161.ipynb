## 房屋户型

We can split the feature into how many bedrooms, kitchens, etc.

```python
import re

pattern_room = re.compile(r'(\d+)室|(\d+)房间')
pattern_hall = re.compile(r'(\d+)厅')
pattern_kitchen = re.compile(r'(\d+)厨')
pattern_bath = re.compile(r'(\d+)卫')

def extract_layout(text, pattern):
    if pd.isna(text) or str(text).strip() == '':
        return pd.NA
    match = pattern.search(str(text))
    if match:
        return int(match.group(1)) if match.group(1) else int(match.group(2))
    return 0

for df_name in ['housing_train', 'housing_test']:
    df = locals()[df_name]
    if '房屋户型' in df.columns:
        df['室'] = df['房屋户型'].apply(lambda x: extract_layout(x, pattern_room))
        df['厅'] = df['房屋户型'].apply(lambda x: extract_layout(x, pattern_hall))
        df['厨'] = df['房屋户型'].apply(lambda x: extract_layout(x, pattern_kitchen))
        df['卫'] = df['房屋户型'].apply(lambda x: extract_layout(x, pattern_bath))
        df = df.drop('房屋户型', axis=1)
        locals()[df_name] = df

for df_name in ['rent_train', 'rent_test']:
    df = locals()[df_name]
    if '户型' in df.columns:
        df['室'] = df['户型'].apply(lambda x: extract_layout(x, pattern_room))
        df['厅'] = df['户型'].apply(lambda x: extract_layout(x, pattern_hall))
        df['厨'] = df['户型'].apply(lambda x: extract_layout(x, pattern_kitchen))
        df['卫'] = df['户型'].apply(lambda x: extract_layout(x, pattern_bath))
        df = df.drop('户型', axis=1)
        locals()[df_name] = df
```

## 房屋楼层

```python
floor_categories = ["地下室", "底层", "低楼层", "中楼层", "高楼层", "顶层"]
floor_category_map = {
    "地下室": 0,
    "底层": 1,
    "低楼层": 2,
    "中楼层": 5,
    "高楼层": 4,
    "顶层": 3
}
```

```python
def process_rent_floor(text):
    if pd.isna(text) or str(text).strip() == "":
        return pd.NA
    for cat in floor_categories:
        if cat in str(text):
            return floor_category_map[cat]
    match = re.match(r'(\d+)/(\d+)层', str(text))
    if match:
        floor, total = int(match.group(1)), int(match.group(2))
        ratio = floor / total
        if ratio < 0.2:
            return floor_category_map["底层"]
        elif ratio < 0.4:
            return floor_category_map["低楼层"]
        elif ratio < 0.6:
            return floor_category_map["中楼层"]
        elif ratio < 0.8:
            return floor_category_map["高楼层"]
        else:
            return floor_category_map["顶层"]
    return pd.NA

datasets = {
    "housing_train": "所在楼层",
    "housing_test": "所在楼层",
    "rent_train": "楼层",
    "rent_test": "楼层"
}

for ds_name, col_name in datasets.items():
    df = locals()[ds_name]
    if col_name in df.columns:
        df['楼高'] = df[col_name].apply(process_rent_floor)
        df = df.drop(col_name, axis=1)
        locals()[ds_name] = df
```

```python
structure_map = {
    "钢结构": 6, "钢混结构": 5, "框架结构": 4, 
    "混合结构": 3, "砖混结构": 2, "砖木结构": 1,
    "未知结构": 3.5
}

decoration_map = {
    "精装": 4, "其他": 3, "毛坯": 2, "简装": 1,
}
```

```python
for ds in ["housing_train", "housing_test"]:
    df = locals()[ds]
    if "建筑结构" in df.columns:
        df["结构"] = df["建筑结构"].map(structure_map).fillna(3.5)
        df = df.drop("建筑结构", axis=1)
    if "装修情况" in df.columns:
        df["装修"] = df["装修情况"].map(decoration_map).fillna(2.5)
        df = df.drop("装修情况", axis=1)
    for col in ["结构", "装修"]:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')
    locals()[ds] = df
```

```python
def chinese_to_arabic(chinese_num):
    if pd.isna(chinese_num) or str(chinese_num).strip() == "":
        return pd.NA
    chinese_digits = {'零':0, '一':1, '二':2, '两':2, '三':3, '四':4, '五':5, '六':6, '七':7, '八':8, '九':9}
    units = {'十':10, '百':100, '千':1000, '万':10000}
    result = 0
    unit = 1
    for char in reversed(str(chinese_num).strip()):
        if char in units:
            unit = units[char]
        elif char in chinese_digits:
            result += chinese_digits[char] * unit
            unit *= 10 if unit < 10 else 1
        else:
            return pd.NA
    return result if result !=0 else 0
```

```python
def process_floor_ratio(df):
    if "梯户比例" not in df.columns:
        return df
    df['梯数_中文'] = df['梯户比例'].apply(lambda x: re.search(r'(.+?)梯', str(x)).group(1) if re.search(r'(.+?)梯', str(x)) else pd.NA)
    df['户数_中文'] = df['梯户比例'].apply(lambda x: re.search(r'梯(.+?)户', str(x)).group(1) if re.search(r'梯(.+?)户', str(x)) else pd.NA)
    df['梯数'] = df['梯数_中文'].apply(chinese_to_arabic)
    df['户数'] = df['户数_中文'].apply(chinese_to_arabic)
    df['梯数'] = pd.to_numeric(df['梯数'], errors='coerce')
    df['户数'] = pd.to_numeric(df['户数'], errors='coerce')
    df['户梯比'] = df.apply(lambda row: round(row['户数']/row['梯数'], 2) if pd.notna(row['梯数']) and pd.notna(row['户数']) and row['梯数'] !=0 else pd.NA, axis=1)
    df['户梯比'] = pd.to_numeric(df['户梯比'], errors='coerce')
    df = df.drop(['梯户比例', '梯数_中文', '户数_中文','梯数','户数'], axis=1)
    return df

for ds in ["housing_train", "housing_test"]:
    df = locals()[ds]
    df = process_floor_ratio(df)
    locals()[ds] = df
```

## 房屋年龄

```python
def calculate_house_age(text):
    if pd.isna(text) or str(text).strip() == "":
        return pd.NA
    text = str(text).strip()
    match = re.search(r'(\d+(-\d+)?)', text)
    if not match:
        return pd.NA
    year_str = match.group(1)
    if '-' in year_str:
        start, end = map(int, year_str.split('-'))
        avg_year = (start + end) / 2
    else:
        avg_year = int(year_str)
    return 2025 - avg_year

for ds in ["housing_train", "housing_test", "rent_train", "rent_test"]:
    df = locals()[ds]
    if "建筑年代" in df.columns:
        df["房屋年龄"] = df["建筑年代"].apply(calculate_house_age)
        df = df.drop("建筑年代", axis=1)
        df['房屋年龄'] = pd.to_numeric(df['房屋年龄'], errors='coerce')
        locals()[ds] = df

print(housing_train["房屋年龄"].head())
print(rent_train["房屋年龄"].head())
```

## 物业费

```python
def process_value(text, target_char):
    if pd.isna(text) or str(text).strip() == "":
        return pd.NA
    text = str(text).strip()
    match = re.search(f'(.+?){target_char}', text)
    if not match:
        return pd.NA
    value_str = match.group(1).strip()
    if '-' in value_str:
        parts = [float(part.strip()) for part in value_str.split('-') if part.strip().replace('.','').isdigit()]
        if len(parts) == 2:
            return (parts[0] + parts[1]) / 2
    elif value_str.replace('.','').isdigit():
        return float(value_str)
    return pd.NA

for ds in ["housing_train", "housing_test", "rent_train", "rent_test"]:
    df = locals()[ds]
    if "物 业 费" in df.columns:
        df = df.rename(columns={"物 业 费": "物业费"})
    if "物业费" in df.columns:
        df["物业费"] = df["物业费"].apply(lambda x: process_value(x, target_char="元"))
        df['物业费'] = pd.to_numeric(df['物业费'], errors='coerce')
    locals()[ds] = df
```

## 燃气费、供电、供水、楼栋总数

```python
def process_gas_fee(text):
    if pd.isna(text) or str(text).strip() == "":
        return pd.NA
    text = str(text).strip()
    match = re.search(r'(.+?)元', text)
    if not match:
        return pd.NA
    value_str = match.group(1).strip()
    if '-' in value_str:
        parts = [float(part.strip()) for part in value_str.split('-') if part.strip().replace('.', '').isdigit()]
        if len(parts) == 2:
            return (parts[0] + parts[1]) / 2
    elif value_str.replace('.', '').isdigit():
        return float(value_str)
    return pd.NA

def process_water_electric(text, type_name):
    if pd.isna(text) or str(text).strip() == "":
        return pd.NA
    text = str(text).strip()
    if type_name == "供水":
        if "民水" in text and "商水" in text:
            return 2
        elif "民水" in text:
            return 0
        elif "商水" in text:
            return 1
    elif type_name == "供电":
        if "民电" in text and "商电" in text:
            return 2
        elif "民电" in text:
            return 0
        elif "商电" in text:
            return 1
    return pd.NA

def process_building_count(text):
    if pd.isna(text) or str(text).strip() == "":
        return pd.NA
    text = str(text).strip()
    match = re.search(r'(\d+)栋', text)
    if match:
        return int(match.group(1))
    return pd.NA
```

```python
datasets = ["housing_train", "housing_test", "rent_train", "rent_test"]
for ds in datasets:
    df = locals()[ds]  

    if "燃气费" in df.columns:
        df["燃气费"] = df["燃气费"].apply(process_gas_fee)
        df["燃气费"] = pd.to_numeric(df["燃气费"], errors="coerce")

    if "供水" in df.columns:
        df["供水类型"] = df["供水"].apply(lambda x: process_water_electric(x, "供水"))
        df["供水类型"] = pd.to_numeric(df["供水类型"], errors="coerce")
        df = df.drop(columns="供水", errors="ignore")  

    if "供电" in df.columns:
        df["供电类型"] = df["供电"].apply(lambda x: process_water_electric(x, "供电"))
        df["供电类型"] = pd.to_numeric(df["供电类型"], errors="coerce")
        df = df.drop(columns="供电", errors="ignore")  

    if "楼栋总数" in df.columns:
        df["楼栋总数"] = df["楼栋总数"].apply(process_building_count)
        df["楼栋总数"] = pd.to_numeric(df["楼栋总数"], errors="coerce")
    
    locals()[ds] = df 
```

```python
def iqr_positive(df, feat):
    q1, q3 = df[feat].quantile([0.25, 0.75])
    iqr = q3 - q1
    lower = max(q1 - 1.5*iqr, 0)  
    upper = q3 + 1.5*iqr
    return lower, upper, df[(df[feat] < lower) | (df[feat] > upper)]

def zscore_positive(df, feat, threshold=3):
    mean, std = df[feat].mean(), df[feat].std()
    lower = max(mean - threshold*std, 0)  
    upper = mean + threshold*std
    z = (df[feat] - mean)/std
    return lower, upper, df[abs(z) > threshold]

def range_01_positive(df, feat):
    lower = 0  
    upper = 1  
    outliers = df[(df[feat] < lower) | (df[feat] > upper)]
    return lower, upper, outliers
    
feature_methods = {
    "建筑面积": iqr_positive, "物业费": iqr_positive, "楼高": iqr_positive, "房屋年龄": iqr_positive,
    "绿化率": range_01_positive, "容积率": iqr_positive, "户梯比": zscore_positive,
    "室": iqr_positive, "厅": iqr_positive, "厨": iqr_positive, "卫": iqr_positive,"面积":iqr_positive,"燃气费":iqr_positive
}
```

```python
for feat, method in feature_methods.items():
    if feat in housing_train.columns:
        df_clean = housing_train[housing_train[feat].notna()].copy()
        non_positive = df_clean[df_clean[feat] <= 0]
        positive_data = df_clean[df_clean[feat] > 0]
        if positive_data.empty:
            lower, upper, extreme = 0, 0, pd.DataFrame()
        else:
            lower, upper, extreme = method(positive_data, feat)
        outlier_idx = pd.concat([non_positive, extreme], ignore_index=True).index
        housing_train.loc[outlier_idx, feat] = pd.NA
        
        print(f"housing_train：{feat}")
        print(f"阈值：<{lower:.2f} 或 >{upper:.2f}")
        print(f"异常值数量：{len(outlier_idx)}\n{'-'*50}")
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.compose import ColumnTransformer

datasets = {
    "housing_train": X_housing_train,
    "rent_train": X_rent_train,
    "housing_test":X_housing_test,
    "rent_test":X_rent_test,
    "housing_predict":housing_test_no_id,
    "rent_predict":rent_test_no_id
}

special_features = ["城市",  "板块"]

def generate_polynomial_features(df):
    df_special = [col for col in special_features if col in df.columns]
    df_ordinary = [col for col in df.columns if col not in df_special]
    
    if not df_special and not df_ordinary:
        return df  

    transformers = []
    if df_special:
        transformers.append(
            ("special_poly", PolynomialFeatures(degree=3, include_bias=False), df_special)
        )
    if df_ordinary:
        transformers.append(
            ("ordinary_poly", PolynomialFeatures(degree=3, include_bias=False), df_ordinary)
        )

    ct = ColumnTransformer(transformers=transformers, remainder="drop")  
    poly_features = ct.fit_transform(df)

    feature_names = []
    if df_special:
        special_poly = ct.named_transformers_["special_poly"]
        feature_names.extend(special_poly.get_feature_names_out(df_special))
    if df_ordinary:
        ordinary_poly = ct.named_transformers_["ordinary_poly"]
        feature_names.extend(ordinary_poly.get_feature_names_out(df_ordinary))

    return pd.DataFrame(poly_features, columns=feature_names, index=df.index)

X_housing_train_poly = generate_polynomial_features(datasets["housing_train"])
X_rent_train_poly = generate_polynomial_features(datasets["rent_train"])
X_housing_test_poly = generate_polynomial_features(datasets["housing_test"])
X_rent_test_poly = generate_polynomial_features(datasets["rent_test"])
```

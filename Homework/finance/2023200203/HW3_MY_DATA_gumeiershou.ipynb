```python
import requests  # 引入requests库，用于发送HTTP请求
from lxml import etree  # 引入lxml库中的etree模块，用于解析HTML文档

# 定义目标URL，即要爬取的网页地址
url='https://sh.esf.fang.com/house-a018-b01601/'

# 定义HTTP请求头，其中包括User-Agent信息，用于伪装成浏览器进行访问
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/141.0.0.0 Safari/537.36'
}

# 发送GET请求获取网页内容，并将响应内容存储在resp变量中
resp = requests.get(url, headers=headers)
# 设置响应内容的编码格式为utf-8，确保中文字符正常显示
resp.encoding = 'utf-8'

#一共有79页，截取前20页数据
#生成数列
sequence = []
# 用一个循环覆盖1-20的范围
for num in range(1, 21):
    # 拼接"3"和当前数字，转换为整数
    value = int(f"3{num}")
    sequence.append(value)

Names = []
Price = []
Average_Price = []
Residual = []
Residual_location = []
Subway = []
Type = []
Area = []
Floor = []
Toward = []
Build_Year = []


    
for i in sequence:
    # 构建每一页的URL
    url = f'https://sh.esf.fang.com/house-a018-b01601/i3{i}/'

    # 发送GET请求获取网页内容，并将响应内容存储在resq变量中
    resq = requests.get(url, headers=headers)
    # 设置响应内容的编码格式为utf-8，确保中文字符正常显示
    resq.encoding = 'utf-8'

    # 使用etree.HTML方法将HTML文本解析为一个HTML文档对象
    e = etree.HTML(resq.text)
    nodes = e.xpath("//p[@class='tel_shop']")
    normal_length = 5
    filled_nodes_data = []  # 用来存填充后的数据列表

    for item in nodes:
        # 先提取节点的文本内容，处理成列表形式
        text = item.xpath('string(.)')
        clean_text = " ".join(text.split())
        parts = clean_text.split("|")
        # 现在 parts 是提取出的各个部分组成的列表
        if len(parts) == normal_length:
            filled_nodes_data.append(parts)
        else:
            filled_item = parts.copy()  # 这里 parts 是普通列表，有 copy 方法
            while len(filled_item) < normal_length:
                filled_item.append('')
            filled_nodes_data.append(filled_item)

    # 然后再从 filled_nodes_data 中提取第一列等数据
    
    for data_list in filled_nodes_data:
        if data_list:  # 确保列表不为空
            column1 = data_list[0].strip()
            Type.append(column1)

    
    for data_list in filled_nodes_data:
        if data_list:  # 确保列表不为空
            column2 = data_list[1].strip()
            Area.append(column2)

    
    for data_list in filled_nodes_data:
        if data_list:  # 确保列表不为空
            column3 = data_list[2].strip()
            Floor.append(column3)

    
    for data_list in filled_nodes_data:
        if data_list:  # 确保列表不为空
            column4 = data_list[3].strip()
            Toward.append(column4)

    
    for data_list in filled_nodes_data:
        if data_list:  # 确保列表不为空
            column5 = data_list[4].strip()
            Build_Year.append(column5)


    # 使用extend()添加提取的数据
    Names.extend([n.xpath("string(.)").strip() for n in e.xpath("//span[@class='tit_shop']")])
    Price.extend([n.xpath("string(.)").strip() for n in e.xpath("//span[@class='red']")])
    Average_Price.extend([n.xpath("string(.)").strip() for n in e.xpath("//dd[@class='price_right']/span[2]")])
    Residual.extend([n.xpath("string(.)").strip() for n in e.xpath("//p[@class='add_shop']/a")])
    Residual_location.extend([n.xpath("string(.)").strip() for n in e.xpath("//p[@class='add_shop']/span")])
    Subway.extend([n.xpath("string(.)").strip() for n in e.xpath("//span[@class='bg_none icon_dt']")])
    
print(Subway)



with open('古美二手房.txt', 'w', encoding='utf-8') as f:
# 使用zip函数组合在一起，逐行写入文件
    for a,b,c,d,e,l,g,h,i,j,k in zip(Names, Price, Average_Price, Type,Area,Floor,Toward,Build_Year,Residual,Residual_location,Subway):
    # 写入格式为：红球号码：xxx 蓝球号码：xxx
        f.write(f'名称:{a} 价格:{b} 每平方米价格:{c} 户型:{d}面积:{e}朝向:{l}楼层:{g}建造年代:{h}小区:{i}小区地址:{j}地铁站:{k}\n')
print('ok')
```

```python
import pandas as pd
import re

# 定义要提取的字段名和对应的匹配关键词（增加灵活性）
fields = [
    {"name": "名称", "keyword": "名称:"},
    {"name": "价格", "keyword": "价格:"},
    {"name": "每平方米价格", "keyword": "每平方米价格:"},
    {"name": "户型", "keyword": "户型:"},
    {"name": "面积", "keyword": "面积:"},
    {"name": "朝向", "keyword": "朝向:"},
    {"name": "楼层", "keyword": "楼层:"},
    {"name": "建造年代", "keyword": ["建造年代:", "年代:"]},  # 支持多个关键词
    {"name": "小区", "keyword": "小区:"},
    {"name": "小区地址", "keyword": "小区地址:"},
    {"name": "地铁站", "keyword": "地铁站:"}
]

all_data = []

with open("古美二手房.txt", "r", encoding="utf-8") as f:
    for line_num, line in enumerate(f, 1):
        line = line.strip()
        if not line:
            continue
        
        data = {field["name"]: "" for field in fields}  # 初始化所有字段为空
        remaining_text = line  # 剩余未匹配的文本
        
        for field in fields:
            # 处理多关键词情况
            keywords = field["keyword"] if isinstance(field["keyword"], list) else [field["keyword"]]
            found = False
            
            for kw in keywords:
                # 查找关键词在剩余文本中的位置
                idx = remaining_text.find(kw)
                if idx != -1:
                    # 提取关键词后的内容（直到下一个关键词或文本结束）
                    content_start = idx + len(kw)
                    # 查找下一个关键词的位置
                    next_kw_pos = len(remaining_text)  # 默认到文本结束
                    for next_field in fields:
                        next_kws = next_field["keyword"] if isinstance(next_field["keyword"], list) else [next_field["keyword"]]
                        for next_kw in next_kws:
                            pos = remaining_text.find(next_kw, content_start)
                            if pos != -1 and pos < next_kw_pos:
                                next_kw_pos = pos
                    # 提取内容
                    content = remaining_text[content_start:next_kw_pos].strip()
                    data[field["name"]] = content
                    found = True
                    break
            
            if found:
                # 更新剩余文本（去掉已匹配的部分）
                remaining_text = remaining_text[:idx] + remaining_text[next_kw_pos:]
        
        # 特殊处理：如果"建造年代"仍为空，尝试从文本中提取年份（如"2002年建"）
        if not data["建造年代"]:
            year_match = re.search(r"(\d{4})年建", line)
            if year_match:
                data["建造年代"] = year_match.group(1) + "年"
        
        all_data.append(data)

# 转换为DataFrame并导出
df = pd.DataFrame(all_data)
df.to_excel("古美二手房3.xlsx", index=False, engine="openpyxl")
print("优化后的Excel已生成：fang_data_improved3.xlsx")
```

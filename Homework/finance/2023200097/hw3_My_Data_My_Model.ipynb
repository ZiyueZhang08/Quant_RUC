```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# 假设你的数据框为df，包含房价和租金数据
# 例如：df = pd.DataFrame({'price_per_m2': [...], 'rent_per_m2': [...], 'location': [...]})

def basic_data_description(df, numerical_columns):
    """
    基本数据描述
    """
    print("=" * 50)
    print("基本数据描述")
    print("=" * 50)
    
    # 数据概览
    print("数据形状:", df.shape)
    print("\n数据类型:")
    print(df.dtypes)
    
    # 数值型变量的描述性统计
    print("\n描述性统计:")
    print(df[numerical_columns].describe())
    
    # 缺失值检查
    print("\n缺失值统计:")
    print(df.isnull().sum())
    
    return df[numerical_columns].describe()

numerical_cols = ['Price', 'Area']
des_df1s = basic_data_description(df1s, numerical_cols )
```

```python
def detect_outliers_iqr(df, column):
    """
    使用IQR方法检测异常值
    """
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    
    print(f"\n{column} - IQR异常值检测:")
    print(f"Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")
    print(f"下限: {lower_bound:.2f}, 上限: {upper_bound:.2f}")
    print(f"异常值数量: {len(outliers)}")
    print(f"异常值比例: {len(outliers)/len(df)*100:.2f}%")
    
    return outliers

def detect_outliers_zscore(df, column, threshold=3):
    """
    使用Z-score方法检测异常值
    """
    z_scores = np.abs(stats.zscore(df[column].dropna()))
    outliers = df[z_scores > threshold]
    
    print(f"\n{column} - Z-score异常值检测 (阈值={threshold}):")
    print(f"异常值数量: {len(outliers)}")
    print(f"异常值比例: {len(outliers)/len(df)*100:.2f}%")
    
    return outliers

def detect_outliers_mad(df, column, threshold=3):
    """
    使用MAD(中位数绝对偏差)方法检测异常值
    """
    median = df[column].median()
    mad = np.median(np.abs(df[column] - median))
    modified_z_scores = 0.6745 * (df[column] - median) / mad
    outliers = df[np.abs(modified_z_scores) > threshold]
    
    print(f"\n{column} - MAD异常值检测 (阈值={threshold}):")
    print(f"异常值数量: {len(outliers)}")
    print(f"异常值比例: {len(outliers)/len(df)*100:.2f}%")
    
    return outliers

def detect_outliers_zscore_simple(df, column, threshold=3):
    """
    使用Z-score方法检测异常值（简化修复版本）
    """
    # 直接处理，不dropna
    z_scores = np.abs(stats.zscore(df[column], nan_policy='omit'))
    
    # 处理NaN值
    outlier_mask = pd.Series(False, index=df.index)
    valid_mask = ~pd.isna(z_scores)
    outlier_mask[valid_mask] = z_scores[valid_mask] > threshold
    
    outliers = df[outlier_mask]
    
    print(f"\n{column} - Z-score异常值检测 (阈值={threshold}):")
    print(f"异常值数量: {len(outliers)}")
    print(f"异常值比例: {len(outliers)/len(df)*100:.2f}%")
    
    return outliers

def detect_outliers_mad_simple(df, column, threshold=3):
    """
    使用MAD方法检测异常值（简化修复版本）
    """
    data = df[column].dropna()
    
    if len(data) == 0:
        print(f"警告: {column} 无有效数据")
        return df.iloc[[]]
    
    median = data.median()
    mad = np.median(np.abs(data - median))
    
    if mad == 0:
        print(f"警告: {column} 的MAD为0，无法计算异常值")
        return df.iloc[[]]
    
    modified_z_scores = 0.6745 * (data - median) / mad
    outlier_indices = data.index[np.abs(modified_z_scores) > threshold]
    outliers = df.loc[outlier_indices]
    
    print(f"\n{column} - MAD异常值检测 (阈值={threshold}):")
    print(f"异常值数量: {len(outliers)}")
    print(f"异常值比例: {len(outliers)/len(df)*100:.2f}%")
    
    return outliers

def comprehensive_outlier_analysis(df, numerical_columns):
    """
    综合异常值分析（修复版本）
    """
    print("=" * 50)
    print("异常值分析报告")
    print("=" * 50)
    
    outlier_report = {}
    
    for column in numerical_columns:
        print(f"\n{'='*30}")
        print(f"分析变量: {column}")
        print(f"{'='*30}")
        
        # 检查列中是否有数据
        if df[column].isna().all():
            print(f"警告: {column} 列全部为缺失值")
            continue
            
        # 各种方法检测异常值
        iqr_outliers = detect_outliers_iqr(df, column)
        zscore_outliers = detect_outliers_zscore_simple(df, column)
        mad_outliers = detect_outliers_mad_simple(df, column)
        
        # 收集结果
        outlier_report[column] = {
            'iqr': iqr_outliers,
            'zscore': zscore_outliers,
            'mad': mad_outliers,
            'iqr_count': len(iqr_outliers),
            'zscore_count': len(zscore_outliers),
            'mad_count': len(mad_outliers)
        }
    
    return outlier_report
def visualize_outliers(df, numerical_columns):
    """
    可视化异常值
    """
    # 箱线图
    plt.figure(figsize=(15, 5))
    
    plt.subplot(1, 3, 1)
    df[numerical_columns].boxplot()
    plt.title('箱线图 - 异常值检测')
    plt.xticks(rotation=45)
    
    # 分布图
    plt.subplot(1, 3, 2)
    for col in numerical_columns:
        sns.histplot(df[col], kde=True, alpha=0.5, label=col)
    plt.title('分布图')
    plt.legend()
    
    # 散点图（如果只有两个主要数值变量）
    if len(numerical_columns) >= 2:
        plt.subplot(1, 3, 3)
        plt.scatter(df[numerical_columns[0]], df[numerical_columns[1]], alpha=0.6)
        plt.xlabel(numerical_columns[0])
        plt.ylabel(numerical_columns[1])
        plt.title('散点图')
    
    plt.tight_layout()
    plt.show()


# 完整的数据描述和异常值检测流程
def complete_data_analysis_simple(df, numerical_columns):
    """
    简化的数据分析流程（不包含按地区绘图）
    """
    # 1. 基本描述
    basic_stats = basic_data_description(df, numerical_columns)
    
    # 2. 异常值检测
    outlier_report = comprehensive_outlier_analysis(df, numerical_columns)
    
    # 3. 可视化（不按地区分组）
    visualize_outliers(df, numerical_columns)
    
    # 4. 生成汇总报告
    print("\n" + "="*50)
    print("异常值检测汇总报告")
    print("="*50)
    
    summary_data = []
    for col in numerical_columns:
        summary_data.append({
            '变量': col,
            'IQR异常值数': outlier_report[col]['iqr_count'],
            'Z-score异常值数': outlier_report[col]['zscore_count'],
            'MAD异常值数': outlier_report[col]['mad_count'],
            '数据总数': len(df)
        })
    
    summary_df = pd.DataFrame(summary_data)
    print(summary_df)
    
    return basic_stats, outlier_report, summary_df

# 使用简化版本
numerical_cols = ['Price', 'Area']
basic_stats, outliers, summary = complete_data_analysis_simple(df1s, numerical_cols)
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False

class SeparateHousingModel:
    def __init__(self):
        self.location_encoder = LabelEncoder()
        self.model_price = None
        self.model_rent = None
    
    def prepare_combined_locations(self, df_sale, df_rent):
        """准备统一的位置编码"""
        # 合并所有位置来学习编码
        all_locations = pd.concat([df_sale['location'], df_rent['location']])
        self.location_encoder.fit(all_locations)
        encoding_dict = dict(zip(self.location_encoder.classes_, range(len(self.location_encoder.classes_))))
        print(f"位置编码: {encoding_dict}")
        return self.location_encoder
    
    def clean_data(self, df):
        """清洗数据，处理缺失值和无穷值"""
        data = df.copy()
        
        # 处理无穷值
        numeric_cols = data.select_dtypes(include=[np.number]).columns
        for col in numeric_cols:
            if col in data.columns:
                data[col] = data[col].replace([np.inf, -np.inf], np.nan)
        
        # 删除包含NaN的行
        data_clean = data.dropna()
        print(f"清洗后数据形状: {data_clean.shape}")
        
        return data_clean
    
    def train_price_model(self, df_sale):
        """训练价格模型"""
        # 清洗数据
        data = self.clean_data(df_sale)
        
        if len(data) == 0:
            raise ValueError("价格数据清洗后为空")
        
        data['location_encoded'] = self.location_encoder.transform(data['location'])
        
        # 准备特征矩阵 - 使用statsmodels的add_constant
        X = data[['Area', 'location_encoded']]
        X = sm.add_constant(X)  # 添加常数项
        y = data['ratio']
        
        print(f"价格模型特征矩阵形状: {X.shape}")
        
        self.model_price = sm.OLS(y, X).fit()
        return self.model_price
    
    def train_rent_model(self, df_rent):
        """训练租金模型"""
        # 清洗数据
        data = self.clean_data(df_rent)
        
        if len(data) == 0:
            raise ValueError("租金数据清洗后为空")
        
        data['location_encoded'] = self.location_encoder.transform(data['location'])
        
        # 准备特征矩阵 - 使用statsmodels的add_constant
        X = data[['Area', 'location_encoded']]
        X = sm.add_constant(X)  # 添加常数项
        y = data['ratio']
        
        print(f"租金模型特征矩阵形状: {X.shape}")
        
        self.model_rent = sm.OLS(y, X).fit()
        return self.model_rent
    
    def predict_for_all_blocks(self, df_sale, df_rent):
        """为所有板块预测价格和租金"""
        # 获取所有唯一位置
        all_blocks = list(set(df_sale['location'].unique()) | set(df_rent['location'].unique()))
        
        predictions = []
        
        for block in all_blocks:
            # 使用该板块的平均面积进行预测
            block_sale_data = df_sale[df_sale['location'] == block]
            block_rent_data = df_rent[df_rent['location'] == block]
            
            # 计算平均面积
            if len(block_sale_data) > 0:
                area = block_sale_data['Area'].mean()
            elif len(block_rent_data) > 0:
                area = block_rent_data['Area'].mean()
            else:
                area = (df_sale['Area'].mean() + df_rent['Area'].mean()) / 2
            
            # 确保面积是有效数值
            if pd.isna(area) or np.isinf(area):
                area = 100  # 默认值
            
            # 准备预测数据 - 修复维度问题
            pred_data = pd.DataFrame({
                'const': [1],  # 手动添加常数项
                'Area': [area],
                'location_encoded': [self.location_encoder.transform([block])[0]]
            })
            
            # 确保列顺序与训练时一致
            X_pred = pred_data[['const', 'Area', 'location_encoded']]
            
            print(f"预测数据形状: {X_pred.shape}")
            print(f"预测数据列: {X_pred.columns.tolist()}")
            
            # 预测
            price_pred = self.model_price.predict(X_pred).iloc[0]
            rent_pred = self.model_rent.predict(X_pred).iloc[0]
            price_rent_ratio = price_pred / rent_pred
            
            predictions.append({
                'block': block,
                'area_used': area,
                'predicted_price_per_m2': price_pred,
                'predicted_rent_per_m2': rent_pred,
                'price_rent_ratio': price_rent_ratio
            })
        
        return pd.DataFrame(predictions)

def main_separate_data(df_sale, df_rent):
    """主函数：处理分开的销售和租金数据"""
    
    print("=== 数据概览 ===")
    print(f"销售数据: {len(df_sale)} 条记录")
    print(f"租金数据: {len(df_rent)} 条记录")
    
    # 初始化模型
    model = SeparateHousingModel()
    
    # 准备统一的位置编码
    print("\n=== 准备位置编码 ===")
    model.prepare_combined_locations(df_sale, df_rent)
    
    # 训练价格模型
    print("\n=== 训练价格模型 ===")
    try:
        price_model = model.train_price_model(df_sale)
        print(f"价格模型 R²: {price_model.rsquared:.4f}")
        print("价格模型系数:")
        for param, value in price_model.params.items():
            print(f"  {param}: {value:.2f}")
    except Exception as e:
        print(f"价格模型训练失败: {e}")
        return None, None, None, None
    
    # 训练租金模型
    print("\n=== 训练租金模型 ===")
    try:
        rent_model = model.train_rent_model(df_rent)
        print(f"租金模型 R²: {rent_model.rsquared:.4f}")
        print("租金模型系数:")
        for param, value in rent_model.params.items():
            print(f"  {param}: {value:.2f}")
    except Exception as e:
        print(f"租金模型训练失败: {e}")
        return None, None, None, None
    
    # 为所有板块预测
    print("\n=== 预测所有板块 ===")
    try:
        predictions_df = model.predict_for_all_blocks(df_sale, df_rent)
        print("预测结果:")
        print(predictions_df)
    except Exception as e:
        print(f"预测失败: {e}")
        return None, None, None, None
    
    return predictions_df, model, price_model, rent_model

def plot_figure_b_separate(predictions_df):
    """绘制图B：各板块房价租金比中位数的条形图"""
    
    if predictions_df is None or len(predictions_df) == 0:
        print("没有预测数据可绘制")
        return None
    
    # 按板块排序
    plot_df = predictions_df.sort_values('price_rent_ratio', ascending=False)
    
    plt.figure(figsize=(12, 8))
    bars = plt.bar(plot_df['block'], plot_df['price_rent_ratio'],
                   color='lightblue', alpha=0.7, edgecolor='navy')
    
    # 添加全球公允价值参考线
    plt.axhline(y=200, color='red', linestyle='--', linewidth=3,
                label='全球公允价值 (200)', alpha=0.8)
    
    # 在柱子上显示数值
    for bar, ratio in zip(bars, plot_df['price_rent_ratio']):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, height + 5,
                f'{ratio:.1f}', ha='center', va='bottom', 
                fontsize=11, fontweight='bold')
    
    # 设置图表属性
    plt.title('图B：各板块房价租金比中位数（模型预测）', 
              fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('板块', fontsize=12)
    plt.ylabel('房价租金比', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', alpha=0.3)
    plt.legend(fontsize=12)
    plt.tight_layout()
    plt.show()
    
    # 输出分析结果
    print("\n" + "="*60)
    print("图B分析结果:")
    print("="*60)
    for _, row in plot_df.iterrows():
        diff = row['price_rent_ratio'] - 200
        status = "高于公允价值" if diff > 0 else "低于公允价值"
        print(f"{row['block']}: {row['price_rent_ratio']:.1f} ({status}, 差异: {diff:+.1f})")
    
    return plot_df

# 运行完整流程
print("开始作业3-4数据科学建模（分开数据版本）...")
print("="*60)

# 运行建模
predictions_df, model, price_model, rent_model = main_separate_data(df_sale, df_rent)

if predictions_df is not None:
    # 绘制图B
    print("\n" + "="*60)
    print("绘制图B...")
    print("="*60)
    result_plot = plot_figure_b_separate(predictions_df)

    # 最终总结
    print("\n" + "="*60)
    print("建模总结")
    print("="*60)
    print(f"使用的销售数据: {len(df_sale)} 条")
    print(f"使用的租金数据: {len(df_rent)} 条")
    print(f"预测板块数量: {len(predictions_df)} 个")
    print(f"价格模型性能 R²: {price_model.rsquared:.4f}")
    print(f"租金模型性能 R²: {rent_model.rsquared:.4f}")
    print(f"房价租金比范围: {predictions_df['price_rent_ratio'].min():.1f} - {predictions_df['price_rent_ratio'].max():.1f}")
else:
    print("建模失败，请检查数据质量")
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False

class AdvancedHousingModel:
    def __init__(self):
        self.location_encoder = LabelEncoder()
        self.model_price_basic = None
        self.model_rent_basic = None
        self.model_price_advanced = None
        self.model_rent_advanced = None
        
    def clean_data(self, df):
        """清洗数据，处理缺失值和异常值"""
        data = df.copy()
        
        print(f"原始数据形状: {data.shape}")
        
        # 处理缺失值
        data = data.dropna()
        print(f"删除缺失值后: {data.shape}")
        
        # 处理无穷值
        numeric_cols = data.select_dtypes(include=[np.number]).columns
        for col in numeric_cols:
            if col in data.columns:
                # 替换无穷值为NaN
                data[col] = data[col].replace([np.inf, -np.inf], np.nan)
                # 删除包含NaN的行
                data = data.dropna()
        
        print(f"最终数据形状: {data.shape}")
        return data
    
    def safe_log_transform(self, x):
        """安全的对数变换，避免log(0)"""
        return np.log1p(np.maximum(x, 0))
    
    def safe_sqrt_transform(self, x):
        """安全的平方根变换，避免负数"""
        return np.sqrt(np.maximum(x, 0))
    
    def prepare_combined_locations(self, df_sale, df_rent):
        """准备统一的位置编码"""
        all_locations = pd.concat([df_sale['location'], df_rent['location']])
        self.location_encoder.fit(all_locations)
        encoding_dict = dict(zip(self.location_encoder.classes_, range(len(self.location_encoder.classes_))))
        print(f"位置编码: {encoding_dict}")
        return self.location_encoder
    
    def create_basic_features(self, df):
        """创建基础特征（模型1和模型2）"""
        data = self.clean_data(df)
        data['location_encoded'] = self.location_encoder.transform(data['location'])
        
        # 基础特征：面积 + 位置
        X_basic = data[['Area', 'location_encoded']].copy()
        X_basic = sm.add_constant(X_basic)
        
        print(f"基础特征矩阵形状: {X_basic.shape}")
        print(f"基础特征是否有NaN: {X_basic.isnull().any().any()}")
        print(f"基础特征是否有inf: {np.isinf(X_basic.values).any()}")
        
        return X_basic, data
    
    def create_advanced_features(self, df):
        """创建高级特征（模型1+和模型2+）"""
        data = self.clean_data(df)
        data['location_encoded'] = self.location_encoder.transform(data['location'])
        
        # 基础特征
        X_advanced = data[['Area', 'location_encoded']].copy()
        
        # 1. 非线性特征（使用安全变换）
        X_advanced['Area_squared'] = X_advanced['Area'] ** 2
        X_advanced['Area_log'] = self.safe_log_transform(X_advanced['Area'])
        X_advanced['Area_sqrt'] = self.safe_sqrt_transform(X_advanced['Area'])
        X_advanced['Area_cubed'] = X_advanced['Area'] ** 3
        
        # 2. 交互特征
        X_advanced['Area_location_interaction'] = X_advanced['Area'] * X_advanced['location_encoded']
        
        # 3. 多项式特征
        X_advanced['Area_poly2'] = X_advanced['Area'] ** 2
        X_advanced['Area_poly3'] = X_advanced['Area'] ** 3
        
        # 检查并处理可能产生的无穷值
        for col in X_advanced.columns:
            X_advanced[col] = X_advanced[col].replace([np.inf, -np.inf], np.nan)
        
        # 删除包含NaN的行
        X_advanced = X_advanced.dropna()
        
        # 确保数据对齐
        data = data.loc[X_advanced.index]
        
        # 添加常数项
        X_advanced = sm.add_constant(X_advanced)
        
        print(f"高级特征矩阵形状: {X_advanced.shape}")
        print(f"高级特征是否有NaN: {X_advanced.isnull().any().any()}")
        print(f"高级特征是否有inf: {np.isinf(X_advanced.values).any()}")
        
        return X_advanced, data
    
    def train_all_models(self, df_sale, df_rent):
        """训练所有模型"""
        print("="*80)
        print("开始训练所有模型")
        print("="*80)
        
        # 训练基础价格模型（模型1）
        print("\n--- 训练模型1: 基础价格模型 ---")
        try:
            X_price_basic, data_price = self.create_basic_features(df_sale)
            y_price = data_price['ratio']
            self.model_price_basic = sm.OLS(y_price, X_price_basic).fit()
            print(f"✅ 模型1 R²: {self.model_price_basic.rsquared:.4f}")
        except Exception as e:
            print(f"❌ 模型1训练失败: {e}")
            return None
        
        # 训练基础租金模型（模型2）
        print("\n--- 训练模型2: 基础租金模型 ---")
        try:
            X_rent_basic, data_rent = self.create_basic_features(df_rent)
            y_rent = data_rent['ratio']
            self.model_rent_basic = sm.OLS(y_rent, X_rent_basic).fit()
            print(f"✅ 模型2 R²: {self.model_rent_basic.rsquared:.4f}")
        except Exception as e:
            print(f"❌ 模型2训练失败: {e}")
            return None
        
        # 训练高级价格模型（模型1+）
        print("\n--- 训练模型1+: 高级价格模型 ---")
        try:
            X_price_advanced, data_price = self.create_advanced_features(df_sale)
            y_price = data_price['ratio']
            self.model_price_advanced = sm.OLS(y_price, X_price_advanced).fit()
            print(f"✅ 模型1+ R²: {self.model_price_advanced.rsquared:.4f}")
        except Exception as e:
            print(f"❌ 模型1+训练失败: {e}")
            return None
        
        # 训练高级租金模型（模型2+）
        print("\n--- 训练模型2+: 高级租金模型 ---")
        try:
            X_rent_advanced, data_rent = self.create_advanced_features(df_rent)
            y_rent = data_rent['ratio']
            self.model_rent_advanced = sm.OLS(y_rent, X_rent_advanced).fit()
            print(f"✅ 模型2+ R²: {self.model_rent_advanced.rsquared:.4f}")
        except Exception as e:
            print(f"❌ 模型2+训练失败: {e}")
            return None
        
        return True
    
    
    def predict_with_advanced_models(self, df_sale, df_rent):
        """使用模型1+和模型2+预测所有数据"""
        print("\n" + "="*80)
        print("使用模型1+和模型2+进行预测")
        print("="*80)
        
        # 合并所有数据用于预测
        all_data = pd.concat([df_sale, df_rent], ignore_index=True)
        all_blocks = all_data['location'].unique()
        
        predictions = []
        
        for block in all_blocks:
            # 使用该板块的平均面积
            block_data = all_data[all_data['location'] == block]
            area = block_data['Area'].mean()
            
            if pd.isna(area) or np.isinf(area):
                area = all_data['Area'].mean()
            
            # 准备高级特征用于预测（使用安全变换）
            location_code = self.location_encoder.transform([block])[0]
            
            pred_data = pd.DataFrame({
                'const': [1],
                'Area': [area],
                'location_encoded': [location_code],
                'Area_squared': [area ** 2],
                'Area_log': [self.safe_log_transform(area)],
                'Area_sqrt': [self.safe_sqrt_transform(area)],
                'Area_cubed': [area ** 3],
                'Area_location_interaction': [area * location_code],
                'Area_poly2': [area ** 2],
                'Area_poly3': [area ** 3]
            })
            
            # 确保列顺序与训练时一致
            feature_columns = ['const', 'Area', 'location_encoded', 'Area_squared', 
                             'Area_log', 'Area_sqrt', 'Area_cubed', 
                             'Area_location_interaction', 'Area_poly2', 'Area_poly3']
            X_pred = pred_data[feature_columns]
            
            # 使用高级模型预测
            try:
                price_pred = self.model_price_advanced.predict(X_pred).iloc[0]
                rent_pred = self.model_rent_advanced.predict(X_pred).iloc[0]
                price_rent_ratio = price_pred / rent_pred
                
                predictions.append({
                    'block': block,
                    'area_used': area,
                    'predicted_price_per_m2': price_pred,
                    'predicted_rent_per_m2': rent_pred,
                    'price_rent_ratio': price_rent_ratio
                })
            except Exception as e:
                print(f"预测板块 {block} 时出错: {e}")
                continue
        
        predictions_df = pd.DataFrame(predictions)
        print(f"预测完成! 共预测 {len(predictions_df)} 个板块")
        print("\n预测结果:")
        print(predictions_df.round(2))
        
        return predictions_df

def plot_figure_c(predictions_df):
    """绘制图C：各板块房价租金比的条形图"""
    print("\n" + "="*80)
    print("绘制图C")
    print("="*80)
    
    if predictions_df is None or len(predictions_df) == 0:
        print("没有预测数据可绘制")
        return None
    
    # 按房价租金比排序
    plot_df = predictions_df.sort_values('price_rent_ratio', ascending=False)
    
    plt.figure(figsize=(14, 8))
    bars = plt.bar(plot_df['block'], plot_df['price_rent_ratio'],
                   color='lightgreen', alpha=0.7, edgecolor='darkgreen')
    
    # 添加全球公允价值参考线
    plt.axhline(y=200, color='red', linestyle='--', linewidth=3,
                label='全球公允价值 (200)', alpha=0.8)
    
    # 在柱子上显示数值
    for bar, ratio in zip(bars, plot_df['price_rent_ratio']):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, height + 5,
                f'{ratio:.1f}', ha='center', va='bottom', 
                fontsize=11, fontweight='bold')
    
    # 设置图表属性
    plt.title('图C：各板块房价租金比（基于模型1+和模型2+预测）', 
              fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('板块', fontsize=12)
    plt.ylabel('房价租金比', fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', alpha=0.3)
    plt.legend(fontsize=12)
    plt.tight_layout()
    plt.show()
    
    # 输出分析结果
    print("\n图C分析结果:")
    print("="*50)
    for _, row in plot_df.iterrows():
        diff = row['price_rent_ratio'] - 200
        status = "高于公允价值" if diff > 0 else "低于公允价值"
        print(f"{row['block']}: {row['price_rent_ratio']:.1f} ({status}, 差异: {diff:+.1f})")
    
    return plot_df

# 主函数
def main_complete_analysis(df_sale, df_rent):
    
    
    # 初始化模型
    model = AdvancedHousingModel()
    
    # 准备位置编码
    model.prepare_combined_locations(df_sale, df_rent)
    
    # 训练所有模型
    success = model.train_all_models(df_sale, df_rent)
    if not success:
        print("❌ 模型训练失败，无法继续分析")
        return None, None, None
    
    # 使用高级模型预测
    predictions_df = model.predict_with_advanced_models(df_sale, df_rent)
    
    # 绘制图C
    plot_figure_c(predictions_df)
    
    return predictions_df, model

# 运行完整分析
print("作业3-5：数据科学建模 Pro Max")
print("="*80)
predictions_df,  model = main_complete_analysis(df_sale, df_rent)

if predictions_df is not None:
  
    print(f"\n📊 最终结果:")
    print(f"  - 模型1+ R²: {model.model_price_advanced.rsquared:.4f}")
    print(f"  - 模型2+ R²: {model.model_rent_advanced.rsquared:.4f}")
    print(f"  - 预测板块数量: {len(predictions_df)}")
    print(f"  - 房价租金比范围: {predictions_df['price_rent_ratio'].min():.1f} - {predictions_df['price_rent_ratio'].max():.1f}")
else:
    print("❌ 分析失败，请检查数据质量")
```

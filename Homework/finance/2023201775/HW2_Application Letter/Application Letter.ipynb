```python
import pandas as pd
from docxtpl import DocxTemplate
import os
from datetime import datetime

def generate_sop_documents():
    
    # 读取数据
    universities_df = pd.read_excel("university name.xls", sheet_name="Sheet1",header=None
                                   )
    universities = universities_df.iloc[:, 0].tolist()
    
    research_df = pd.read_excel("areas.xls", sheet_name="Sheet1")
    
    # 创建输出目录
    output = "MS_Word_document"
    if not os.path.exists(output):
        os.makedirs(output)
    
    # 循环生成文档
    document_count = 0
    
    for i, university in enumerate(universities):
        for j, research_row in research_df.iterrows():
            area = research_row['Area']
            
            # 准备模板上下文
            context = {
                'University_name': university,
                'Area': area,
                'Journal1': research_row['Journal1'],
                'Journal2': research_row['Journal2'], 
                'Journal3': research_row['Journal3'],
                'Skill1': research_row['Skill1'],
                'Skill2': research_row['Skill2'],
                'Skill3': research_row['Skill3'],
            }
            
            # 生成Word文档
            safe_university = "".join(c for c in university if c.isalnum() or c in (' ', '-', '_')).rstrip()
            safe_area = area.replace(' ', '_')
            output_filename = f"SOP_{safe_university[:30]}_{safe_area}.docx"
            output_path = os.path.join(output, output_filename)
            
            try:
                doc = DocxTemplate("template.docx")
                doc.render(context)
                doc.save(output_path)
                document_count += 1
                print(f"({document_count}/90) 成功生成: {output_filename}")
            except Exception as e:
                print(f"生成 {output_filename} 时出错: {str(e)}")
    
    print(f"\n完成！成功生成了 {document_count} 个文档，保存在 '{output}' 文件夹中")

# 运行主函数
generate_sop_documents()
```

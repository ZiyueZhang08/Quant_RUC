```python
import os
import pandas as pd
from docxtpl import DocxTemplate
from docx2pdf import convert

# 模板路径
template_path = "D:/1 大四上/Python人工智能/作业/HW3/template.docx"

# 导入Excel
df_univ = pd.read_excel("D:/1 大四上/Python人工智能/作业/HW3/University.xlsx")
df_area = pd.read_excel("D:/1 大四上/Python人工智能/作业/HW3/Area.xlsx")

# 指定Word和PDF文件夹
word_folder = r"D:/1 大四上/Python人工智能/作业/HW3/WORD"
pdf_folder = r"D:/1 大四上/Python人工智能/作业/HW3/PDF"

# 创建文件夹（如果不存在）
os.makedirs(word_folder, exist_ok=True)
os.makedirs(pdf_folder, exist_ok=True)

# Word保存函数
def save_word(context, filename, path = template_path):
    try:
        doc = DocxTemplate(path)
        doc.render(context)
        doc.save(filename)
        return True
    except Exception as e:
        print(f"error: {e}")
        return False

# 生成Word文档
success_count = 0
for index1, row1 in df_univ.iterrows():
    university = row1['Universities']   
    
    for index2, row2 in df_area.iterrows():
        context = {
            'name': "Ziyi Liu",
            'university': university,
            'area': row2['Area'],
            'journal': row2['Journal'],
            'skill': row2['Skill']
        }
        
        filename = f"{word_folder}/ApplicationLetter_{university.replace(' ', '_')}_{context['area'].replace(' ', '_')}.docx"
        
        if save_word(context, filename):
            success_count += 1
            print(f"Application Letter for {context['area']} at {university} has been saved")
        else:
            print(f"Failed to save Application Letter for {context['area']} at {university}")

print(f"\n成功生成 {success_count} 份申请信")

# 将Word转为PDF
try:
    convert(word_folder, pdf_folder)
    print(f"PDF转换成功，文件保存到: {pdf_folder}")
except Exception as e:
    print(f"PDF转换失败，错误: {e}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate x values
x = np.linspace(0, 10, 100)

# Create curved data with some noise
y = 2 * np.sin(x) + x/2 + np.random.normal(0, 0.5, 100)

# Reshape x for sklearn
X = x.reshape(-1, 1)

# Fit linear regression
reg = LinearRegression().fit(X, y)
y_pred = reg.predict(X)

# Plot original points and fitted line
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', alpha=0.5, label='Data points')
plt.plot(x, y_pred, color='red', label='Fitted line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Regression Fit to Curved Data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

# Create polynomial features
poly = PolynomialFeatures(degree=4)
X_poly = poly.fit_transform(X)

# Fit linear regression on polynomial features
reg_poly = LinearRegression().fit(X_poly, y)
y_poly_pred = reg_poly.predict(X_poly)

# Plot original points and fitted polynomial line
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', alpha=0.5, label='Data points')
plt.plot(x, y_poly_pred, color='green', label='Fitted polynomial line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Polynomial Regression Fit to Curved Data')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.preprocessing import SplineTransformer

# Create a SplineTransformer with a specified number of knots
spline = SplineTransformer(n_knots=5, degree=1)

# Transform the original x values to include spline features
X_spline = spline.fit_transform(X)

# Fit linear regression on spline features
reg_spline = LinearRegression().fit(X_spline, y)
y_spline_pred = reg_spline.predict(X_spline)

# Plot original points and fitted spline line
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', alpha=0.5, label='Data points')
plt.plot(x, y_spline_pred, color='orange', label='Fitted spline line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Spline Regression Fit to Curved Data')
plt.legend()
plt.grid(True)
plt.show()

```
